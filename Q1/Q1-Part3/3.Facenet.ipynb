Making A Facenet Instance

```python
from keras_facenet import FaceNet
facenet_model = FaceNet()
```

# Applying Facenet On Cropped Faces

```python
!unzip /content/cropped_faces.zip
```

Reading Cropped Faces Images' Paths

```python
main_folder_path = "/content/cropped_faces"
image_paths = []
label = []
for folder in os.listdir(main_folder_path):
    category_path = os.path.join(main_folder_path, folder)
    for image_name in os.listdir(category_path):
        image_path = os.path.join(category_path, image_name)
        image_paths.append(image_path)
        label.append(folder)
```

Getting Embedding Of Each Face By Facenet

```python
def get_embedding(model, face):
    # transfer face into one sample (3 dimension to 4 dimension)
    print(face.shape)
    sample = np.expand_dims(face, axis=0)
    print(sample.shape)
    # make prediction to get embedding
    yhat = model.embeddings(sample)
    return yhat[0]
```

```python
# convert each face in the train set into embedding
emdTrainX = list()
for image_path in image_paths:
    face = cv2.imread(image_path)
    face = cv2.resize(face,(160,160))   #Resizing Each Image to 160*160 because Original Facenet was trained on 160*160 images 
    emd = get_embedding(facenet_model, face) #each embedding is shape of 512 dimension
    emdTrainX.append(emd)
emdTrainX = np.asarray(emdTrainX)
print(emdTrainX.shape)
```

Training A Linear SVM On This Embedded Features & Checking Test Accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder


# Encode class labels using LabelEncoder
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(label)

# Split the data into a training set and a test set (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(emdTrainX, encoded_labels, test_size=0.2, random_state=42)

# Initialize the SVC classifier (you can adjust hyperparameters as needed)
classifier = SVC(kernel='linear', C=1.0, random_state=42)

# Train the classifier on the training data
classifier.fit(X_train, y_train)

# Predict labels for the test set
y_pred = classifier.predict(X_test)

# Calculate accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)

print("Test Accuracy:", accuracy)
```
